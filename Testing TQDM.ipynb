```python
#import libraries

from tqdm.notebook import tqdm
import libtorrent as lt
import time
```

```python
#insert magnet
magnet_link = input ('Enter magnet link:')
```

```python
#create session and add magnet link

ses = lt.session()
ses.listen_on(6881 ,6891)
params = { 'save_path': '/content/drive/MyDrive/Colabs/Test','storage_mode':
lt.storage_mode_t.storage_mode_sparse,}

h= lt.add_magnet_uri(ses,magnet_link,params)


# Wait for metadata to be downloaded
print('Downloading metadata...')
while not h.has_metadata(): 
  time.sleep(1)

# Get torrent info
info = h.get_torrent_info()

# Display list of files and allow user to select files to download

print('Files in torrent:')
for i, file in enumerate(info.files()): 
  print(f'{i}: {file.path}')

selected_files = input('Enter the indices of the files you want to download (separated by spaces): ')
selected_files = [int(i) for i in selected_files.split()]

# Set file priorities
for i, file in enumerate(info.files()): 
  if i in selected_files: h.file_priority(i, 1) 
  else: h.file_priority(i, 0)







```

```python
# Create tqdm progress bar with custom unit and scale
pbar = tqdm(total=100, unit='B', unit_scale=True, unit_divisor=1024)

# Start downloading and update progress bar
print('Starting download...') 
while not h.is_seed(): 
  s = h.status()  
  pbar.set_postfix_str(f'{s.download_rate / 1024 / 1024:.2f} MB/s') 
  pbar.update(s.progress * 100 - pbar.n) 
  time.sleep(1)


pbar.close()
print('Download complete!')
```
